The circumplex model of emotion was developed by James Russell. This model suggests that emotions are distributed in a two-dimensional circular space, containing arousal and valence dimensions.
Arousal represents the vertical axis and valence represents the horizontal axis, while the center of the circle represents a neutral valence and a medium level of arousal. In this model, emotional states can be represented at any level of valence and arousal, or at a neutral level of one or both of these factors.

--> source: Wikipedia haha

##### Change dtype to datetime in time variable

```python
df_data.time = pd.to_datetime(df_data.time, format='%Y-%m-%d %H:%M:%S.%f')
df_data.info()
```

```python
agg_data= pd.DataFrame(df_data.groupby([pd.Grouper(key='time', freq='D'),'variable','id']).value.mean())
agg_df=agg_data.reset_index()
agg_df.head(15)
```

```python
multi = agg_df.set_index(['id', 'time'])
multi
```

```python
pivoted = agg_df.pivot_table(index = ['id','time'],
                         columns = 'variable',values='value',
                         aggfunc = 'sum',
                         #margins = True, # total column
                         ).sort_index()
```

```python
pivoted
```

```python
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('tab20')

num=1
f= plt.figure(figsize=(40,20))
for var in list(pivoted.columns.array):
    ax = f.add_subplot(5,4,num)
    a = np.asarray(pivoted[var].dropna())
    ax.hist(a, bins=70, color=palette(num))
    ax.set_title(var)
    num+=1
```
